```python
import numpy as np
import glob

def loadVelodynePoints(directory):
    arr = []
    
    # Check if the last charact of the given directory is /, if not, append it.
    if (directory[-1] != '/'):
        directory += '/' 
    
    # Load all binary files in order.
    for file in sorted(glob.glob( (directory + '*.bin') )):
        arr.append(np.fromfile(file, dtype=np.float16)) # Loading with float16(half precision) than float32 reduces the ram usage a lot.
        
    return arr
```

```python
data = loadVelodynePoints('./data/velodyne_points/')
```
